# Web scraping player statistics - La Liga

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

urls = {
    'shooting': 'https://fbref.com/en/comps/12/2023-2024/shooting/2023-2024-La-Liga-Stats',
    'passing': 'https://fbref.com/en/comps/12/2023-2024/passing/2023-2024-La-Liga-Stats',
    'gca': 'https://fbref.com/en/comps/12/2023-2024/gca/2023-2024-La-Liga-Stats',
    'defense': 'https://fbref.com/en/comps/12/2023-2024/defense/2023-2024-La-Liga-Stats',
}

for stat, url in urls.items():

    driver.get(url)

    table = driver.find_element(by='id', value='div_stats_' + stat)

    html_content = table.get_attribute('outerHTML')

    df = pd.read_html(html_content)[0]

    df.columns = ['_'.join(col).strip() if 'Unnamed' not in col[0] else col[1] for col in df.columns]

    if 'Nation' in df.columns:
        df['Nation'] = df['Nation'].str.findall(r'[A-Z]+').str.join('')

    df.to_csv(f'la_liga_{stat}_stats.csv', index=False)

driver.quit()
```


# Web scraping market values of players (all leagues)

We decided to scrape the values of players from all of the top-5 leagues (not only EPL and La Liga) to hopefully increase some of the matches, when merging player statistics and their market values. This is because there might be players, who represented clubs in EPL and La Liga in 2023-20204 season, but then moved to a different league (like Bundesliga, for example). 

```python
#Transfer values - EPL

base_url = 'https://www.transfermarkt.us/premier-league/marktwertaenderungen/wettbewerb/GB1/verein_id//land_id/0/pos//detailpos//plus//galerie/0/page/'

all_data = []

num_pages = 21 

for page in range(1, num_pages + 1):
   
    url = base_url + str(page)

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'
    }
    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.content, 'html.parser')

    table = soup.find('div', class_='responsive-table')

    if table:

        rows = table.find_all('tr', class_=['odd', 'even'])

        for row in rows:

            columns = row.find_all('td')
            row_data = [col.text.strip() for col in columns]

            row_data = [data for data in row_data if data]

            if row_data:
                all_data.append(row_data)

columns = ['Player', 'Position', 'Club', 'Market Value', 'Change in Value']

market_values_eng = pd.DataFrame(all_data, columns=columns)

market_values_eng.rename(columns={'Position': 'Name', 'Club': 'Position'}, inplace=True)

market_values_eng.drop(columns=['Player', 'Change in Value'], inplace=True)

market_values_eng.to_csv('/Users/msyzdykova/Desktop/dataset DS project/data/values_eng.csv')
```

# EPL Statistics

```python
#EPL stats
shooting_eng=pd.read_csv('/Users/msyzdykova/Desktop/dataset DS project/eng/shooting.csv')
passing_eng=pd.read_csv('/Users/msyzdykova/Desktop/dataset DS project/eng/passing.csv')
gca_eng=pd.read_csv('/Users/msyzdykova/Desktop/dataset DS project/eng/gca.csv')
defense_eng=pd.read_csv('/Users/msyzdykova/Desktop/dataset DS project/eng/defense.csv')
misc_eng=pd.read_csv('/Users/msyzdykova/Desktop/dataset DS project/eng/misc.csv')
possession_eng=pd.read_csv('/Users/msyzdykova/Desktop/dataset DS project/eng/possession.csv')
```

```python
for i in [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]:
    print(i.columns)
```

```python
#Dropping unnecessary columns
for i in [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]:
    i.drop(columns=['Unnamed: 0', 'Rk', 'Born', 'Matches'], inplace=True)
```

```python
#Dealing with duplicates
#There are duplicates in our dataset from: the way data is organized on the website:
#it contains dividers between groups of players, so we need to drop all of them
dfs = [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]

for i, df in enumerate(dfs):
    df = df[df['Player'] != 'Player']

    dfs[i] = df
    
shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng = dfs
```

```python
for i in [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]:
    print(i.dtypes)
```

```python
#Reformatting numerical columns from string to float
dfs = [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]

for i, df in enumerate(dfs):

    numeric_cols = df.columns[4:]
    
    for col in numeric_cols:
        if df[col].dtype == 'object':
            if col.endswith('%'):
                df[col] = df[col].str.replace('%', '').astype(float) / 100
            else:
                df[col] = pd.to_numeric(df[col], errors='coerce')

    dfs[i] = df

shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng = dfs
```

```python
#Random checks
print(passing_eng['90s'].max())
print(shooting_eng['Standard_SoT%'])
```

```python
#Check for NaN values
for i in [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]:
    print(i.isna().sum())
```

```python
#Check NaN values in shooting stats

#Check distance in shooting stats
#Since the players that have NaN values for Standard_G/SoT, in general, have no other stats (they're all 0), we'll set this column to 0 too
print(shooting_eng[shooting_eng['Standard_Dist'].isna()].iloc[:,6:].sum(axis=1).sum())

#Check Standard_SoT%, Standard_G/Sh, Standard_G/SoT
#These columns have NaN values because the values that they are divided by 0, which in undefined -> so we feel justified to set these values 0
print(shooting_eng[shooting_eng['Standard_SoT%'].isna()]['Standard_SoT'].sum())
print(shooting_eng[shooting_eng['Standard_G/Sh'].isna()]['Standard_Sh'].sum())
print(shooting_eng[shooting_eng['Standard_G/SoT'].isna()]['Standard_SoT'].sum())
```

```python
#Check passing stats 
#All of the values are NaN becaue they are divided by a 0, so we can impute NaN with 0
for i in ['Total_Cmp%', 'Short_Cmp%', 'Medium_Cmp%', 'Long_Cmp%']:
    print(passing_eng[passing_eng[i].isna()][[i, i.replace('%', '')]])
```

```python
#Check possession stats
#All of the values are NaN because they are divided by a 0, so we can impute NaN with 0
print(possession_eng[possession_eng['Take-Ons_Succ%'].isna()]['Take-Ons_Succ'].sum())
print(possession_eng[possession_eng['Take-Ons_Tkld%'].isna()]['Take-Ons_Tkld'].sum())
```

```python
#Check miscallenuous stats
#All of the values are NaN because they are divided by a 0, so we can impute NaN with 0
misc_eng[misc_eng['Aerial Duels_Won%'].isna()]['Aerial Duels_Won'].sum()
```

```python
#we also have some players that are actually duplicated in our dataset
#this occurs because there are players who left one of the clubs in EPL and joined another one
shooting_eng[shooting_eng['Player'].duplicated()]
```

```python
#Dealing with these duplicates
dfs = [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]

for i, df in enumerate(dfs):
    
    #now, these are duplicates, when one player left one of the clubs in EPL and joined another one
    df = df.sort_values(by='90s', ascending=False)

    #instead of dropping these players from our dataset completely, we choose the entry that has the highest 90s
    #meaning he played more matched for that club
    df = df.drop_duplicates(subset='Player', keep='first')

    dfs[i] = df
    
shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng = dfs
```

```python
for i in [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]:
    print(len(i))
```

```python
#We don't consider goalkeepers because their valuation and statistics are very different from on-field players'

dfs = {'shooting_eng': shooting_eng, 'passing_eng': passing_eng, 'gca_eng': gca_eng, 'defense_eng': defense_eng, 'possession_eng': possession_eng, 'misc_eng': misc_eng}

for name, df in dfs.items():
    dfs[name] = df[df['Pos'] != 'GK']

shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng = dfs.values()
```

```python
for i in [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]:
    print(len(i))
```

```python
#We're cleaning column Nation
for df in [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]:
    df['Nation']=df['Nation'].str.findall(r'[A-Z]+').str.join('')
```

```python
#Merged statistics - EPL
dfs = [shooting_eng, passing_eng, gca_eng, defense_eng, possession_eng, misc_eng]

merge_cols = ['Player', 'Nation', 'Pos', 'Squad', 'Age', '90s']

merged_df_eng = dfs[0]

for df in dfs[1:]:
    merged_df_eng = pd.merge(merged_df_eng, df, on=merge_cols, how='outer', suffixes=('', '_dup'))

merged_df_eng = merged_df_eng.loc[:, ~merged_df_eng.columns.str.endswith('_dup')]
```

# Merging player statistics with their market values

```python
final_df=eng_spa_stats.merge(stacked_df, left_on='Player', right_on='Name')
```

```python
final_df
```

```python
#Dropping unnecessary columns because we already have another 2 columns that contain this data
final_df.drop(columns=['Name', 'Position'], inplace=True)
```

```python
#Cleaning market value column
def convert_market_value(value):
    value = value.replace('€', '')  # Remove the euro symbol
    if 'm' in value:
        return float(value.replace('m', '')) * 1e6  # Convert to millions
    elif 'k' in value:
        return float(value.replace('k', '')) * 1e3  # Convert to thousands
    else:
        return float(value)  # Handle values without suffix

final_df['Market Value'] = final_df['Market Value'].apply(convert_market_value)
```

```python
final_df
```

```python
#Cleaning up position column because we need to make sure position is a single value, not multiple
#The 2nd position after the comma in some of the entries illustrated a 2nd priority position of the player or rather a position which he sometimes plays in (but it's not his primary position)
#Hence, we remove the 2nd position after the comma
final_df['Pos'] = final_df['Pos'].apply(lambda x: x.split(',')[0])
```

```python
final_df
```

```python
#Now that we have a combined final dataset we can impute the NaN values with 0
#We are justified to do so because we did check all of the NaN value occurences above and analyzed those situations
final_df.fillna(0, inplace=True)
```
